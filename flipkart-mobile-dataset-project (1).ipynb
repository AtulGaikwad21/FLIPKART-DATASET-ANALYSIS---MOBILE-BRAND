```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python


# Load the dataset
file_path = '/kaggle/input/filpkart-mobiles/Mobiles_Dataset.csv'
data = pd.read_csv(file_path)
data
```

```python
# Display the first few rows of the dataset
data.head(), data.info(), data.describe()
```

```python
# Replace "NIL" and empty strings/spaces with NaN in the 'Actual price' and 'Discount price' columns
data['Actual price'] = data['Actual price'].replace(['NIL', ''], np.nan).replace(r'^\s*$', np.nan, regex=True)
data['Discount price'] = data['Discount price'].replace(['NIL', ''], np.nan).replace(r'^\s*$', np.nan, regex=True)

# Remove any currency symbols and commas, then convert the columns to numeric
data['Actual price'] = data['Actual price'].replace('[\₹,]', '', regex=True).astype(float)
data['Discount price'] = data['Discount price'].replace('[\₹,]', '', regex=True).astype(float)

# Clean the Rating and Reviews columns
data['Rating'] = data['Rating'].replace('[\D]', '', regex=True).astype(int)
data['Reviews'] = data['Reviews'].replace('[\D]', '', regex=True).astype(int)

# Handle missing values in the Camera and RAM (GB) columns without chaining
data['Camera'] = data['Camera'].fillna('Unknown')
data['RAM (GB)'] = data['RAM (GB)'].replace('NIL', '0').astype(int)

# Display the cleaned dataset summary
data.info(), data.describe()
```

**Let's start with the exploratory analysis and visualizations**

```python
import matplotlib.pyplot as plt

# Visualization 1: Distribution of Actual Price and Discount Price
plt.figure(figsize=(12, 6))
plt.hist(data['Actual price'].dropna(), bins=30, alpha=0.5, label='Actual Price')
plt.hist(data['Discount price'].dropna(), bins=30, alpha=0.5, label='Discount Price')
plt.title('Distribution of Actual Price and Discount Price')
plt.xlabel('Price (INR)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

# Visualization 2: Distribution of Ratings
plt.figure(figsize=(10, 6))
plt.hist(data['Stars'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating (Stars)')
plt.ylabel('Number of Products')
plt.grid(True)
plt.show()


# Visualization 3: Scatter Plot - Price vs. Rating
plt.figure(figsize=(10, 6))
plt.scatter(data['Discount price'], data['Stars'], alpha=0.6, edgecolor='w', linewidth=0.5)
plt.title('Price vs. Rating')
plt.xlabel('Discount Price (INR)')
plt.ylabel('Rating (Stars)')
plt.grid(True)
plt.show()
```

```python
# Extracting brand names from product names (assuming the first word is the brand)
data['Brand'] = data['Product Name'].apply(lambda x: x.split()[0])

# Question 1: Best Value for Money (Best Ratings at Lower Prices)
# Calculating the average rating and average discounted price for each brand
brand_value = data.groupby('Brand').agg({'Discount price': 'mean', 'Stars': 'mean'}).sort_values(by='Stars', ascending=False)

# Question 2: Correlation between Display Size and Price
# Calculating the correlation
display_price_correlation = data['Display Size (inch)'].corr(data['Discount price'])

# Question 3: Distribution of Camera Specifications
# Extracting the primary camera resolution (assuming it's the first number before 'MP')
data['Primary Camera (MP)'] = data['Camera'].str.extract(r'(\d+MP)').fillna('Unknown')
camera_distribution = data['Primary Camera (MP)'].value_counts()

# Question 4: Average Rating by RAM Size
ram_rating = data.groupby('RAM (GB)').agg({'Stars': 'mean'}).sort_values(by='Stars', ascending=False)

# Displaying the results
print("Brand Value Analysis:")
print(brand_value.head())

print("\nCorrelation between Display Size and Price:")
print(display_price_correlation)

print("\nCamera Distribution:")
print(camera_distribution.head())

print("\nAverage Rating by RAM Size:")
print(ram_rating.head())
```

```python
# Calculate additional features
# Calculating the discount percentage
data['Discount Percentage'] = ((data['Actual price'] - data['Discount price']) / data['Actual price']) * 100

# Visualization 1: Brand-wise Distribution of Products
plt.figure(figsize=(14, 6))
data['Brand'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Number of Products by Brand')
plt.xlabel('Brand')
plt.ylabel('Number of Products')
plt.grid(True)
plt.show()

# Visualization 2: Average Discount Percentage by Brand
plt.figure(figsize=(14, 6))
brand_discount = data.groupby('Brand')['Discount Percentage'].mean().sort_values(ascending=False)
brand_discount.plot(kind='bar', color='salmon')
plt.title('Average Discount Percentage by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Discount Percentage (%)')
plt.grid(True)
plt.show()

# Visualization 3: RAM Size vs. Discount Price
plt.figure(figsize=(10, 6))
plt.scatter(data['RAM (GB)'], data['Discount price'], alpha=0.7, edgecolor='black')
plt.title('RAM Size vs. Discount Price')
plt.xlabel('RAM (GB)')
plt.ylabel('Discount Price (INR)')
plt.grid(True)
plt.show()

# Visualization 4: Camera Resolution vs. Price
plt.figure(figsize=(14, 6))
camera_price = data[data['Primary Camera (MP)'] != 'Unknown']
camera_price['Primary Camera (MP)'] = camera_price['Primary Camera (MP)'].str.extract(r'(\d+)').astype(int)
camera_price.boxplot(column='Discount price', by='Primary Camera (MP)', grid=True, patch_artist=True)
plt.title('Camera Resolution vs. Price')
plt.suptitle('')  # Suppress the default title
plt.xlabel('Primary Camera (MP)')
plt.ylabel('Discount Price (INR)')
plt.grid(True)
plt.show()
```

```python
# Visualization 1: Price Trends by Brand
plt.figure(figsize=(14, 6))
brand_price_trend = data.groupby('Brand')['Discount price'].mean().sort_values(ascending=False)
brand_price_trend.plot(kind='bar', color='teal')
plt.title('Average Discount Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Discount Price (INR)')
plt.grid(True)
plt.show()

# Visualization 2: Price Trends by RAM Size
plt.figure(figsize=(10, 6))
ram_price_trend = data.groupby('RAM (GB)')['Discount price'].mean().sort_values(ascending=False)
ram_price_trend.plot(kind='bar', color='purple')
plt.title('Average Discount Price by RAM Size')
plt.xlabel('RAM (GB)')
plt.ylabel('Average Discount Price (INR)')
plt.grid(True)
plt.show()

# Visualization 3: Price Trends by Camera Resolution
plt.figure(figsize=(14, 6))
camera_price_trend = camera_price.groupby('Primary Camera (MP)')['Discount price'].mean().sort_values(ascending=False)
camera_price_trend.plot(kind='bar', color='orange')
plt.title('Average Discount Price by Camera Resolution')
plt.xlabel('Primary Camera (MP)')
plt.ylabel('Average Discount Price (INR)')
plt.grid(True)
plt.show()
```

```python
# Filter data for specific brands: Apple, Samsung, and OnePlus
specific_brands = data[data['Brand'].isin(['Apple', 'Samsung', 'OnePlus'])]

# Visualization 1: Price Comparison
plt.figure(figsize=(10, 6))
specific_brands.groupby('Brand')['Discount price'].mean().plot(kind='bar', color=['red', 'blue', 'green'])
plt.title('Average Discount Price by Brand (Apple, Samsung, OnePlus)')
plt.xlabel('Brand')
plt.ylabel('Average Discount Price (INR)')
plt.grid(True)
plt.show()

# Visualization 2: Rating Comparison
plt.figure(figsize=(10, 6))
specific_brands.groupby('Brand')['Stars'].mean().plot(kind='bar', color=['red', 'blue', 'green'])
plt.title('Average Rating by Brand (Apple, Samsung, OnePlus)')
plt.xlabel('Brand')
plt.ylabel('Average Rating (Stars)')
plt.grid(True)
plt.show()

# Visualization 3: Discount Percentage Comparison
plt.figure(figsize=(10, 6))
specific_brands.groupby('Brand')['Discount Percentage'].mean().plot(kind='bar', color=['red', 'blue', 'green'])
plt.title('Average Discount Percentage by Brand (Apple, Samsung, OnePlus)')
plt.xlabel('Brand')
plt.ylabel('Average Discount Percentage (%)')
plt.grid(True)
plt.show()

# Visualization 4: RAM Comparison
plt.figure(figsize=(10, 6))
specific_brands.groupby('Brand')['RAM (GB)'].mean().plot(kind='bar', color=['red', 'blue', 'green'])
plt.title('Average RAM Size by Brand (Apple, Samsung, OnePlus)')
plt.xlabel('Brand')
plt.ylabel('Average RAM Size (GB)')
plt.grid(True)
plt.show()

# Visualization 5: Camera Specification Comparison
plt.figure(figsize=(14, 6))
specific_brands['Primary Camera (MP)'] = specific_brands['Primary Camera (MP)'].str.extract(r'(\d+)').astype(float)
camera_comparison = specific_brands.groupby('Brand')['Primary Camera (MP)'].mean().plot(kind='bar', color=['red', 'blue', 'green'])
plt.title('Average Camera Resolution by Brand (Apple, Samsung, OnePlus)')
plt.xlabel('Brand')
plt.ylabel('Average Camera Resolution (MP)')
plt.grid(True)
plt.show()
```

# Project Summary
Objective:
The project analyzes a dataset of mobile phones from Flipkart, focusing on features such as price, brand, RAM, camera resolution, display size, and user ratings.

**Key Analysis Conducted:
Data Cleaning:**

The dataset required cleaning, particularly with handling non-numeric values in the Actual price and Discount price columns.
Missing values in camera specifications and RAM were managed effectively.
Feature Extraction:

Brand names were extracted from product names to allow for brand-wise analysis.
Ratings Analysis:

The analysis identified top-rated brands, providing insights into which brands are most favored by users.
Average ratings were calculated for different display sizes, RAM sizes, and camera resolutions, revealing trends in user preferences.
Correlation Studies:

The correlation between features like display size and price was examined, showing how certain features might influence pricing.
Findings:
Top Brands: Brands like Apple, OnePlus, and Samsung emerged as top-rated, indicating high user satisfaction.
Feature Impact: Higher RAM and better camera resolutions generally correspond to higher prices and ratings, although correlations are moderate.
Consumer Preferences: The analysis provides insights into consumer preferences based on ratings, which could be valuable for marketing and product development strategies.
